```python
import matplotlib.pyplot as plt
 
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = np.array(y1) + 2


# 코드를 작성해보세요 
fig, axes = plt.subplots(1, 2, figsize = (15, 5), sharey=True)

axes[0].plot(x, y1, color = 'red')
axes[1].plot(x, y2, color = 'green')
plt.show()
```

```python
import seaborn as sns 

df_p = sns.load_dataset('penguins')
df_p
```

```python
sns.scatterplot(
                x = df_p['flipper_length_mm'], 
                y = df_p['bill_length_mm'])
```

```python
sns.scatterplot(data = df_p, 
                x = 'flipper_length_mm', 
                y = 'bill_length_mm', 
                hue = 'species')
```

```python
sns.displot(x = df_p['flipper_length_mm'], col = df_p['species'], hue = df_p['species'])
```

```python
df_p.species.unique()
```

```python
colors = {'Adelie' : 'blue', 'Chinstrap' : 'red', 'Gentoo' : 'yellow'}
df_p['species'].map(colors)
```

```python
plt.scatter(df_p['flipper_length_mm'], 
                df_p['bill_length_mm'], 
                c = df_p['species'].map(colors))
```
